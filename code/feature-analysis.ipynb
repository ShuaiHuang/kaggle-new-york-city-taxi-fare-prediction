# Feature Analysis

This notebook used for analysis the distribution of designed features.

```python
import pandas as pd
from pathlib import Path
from matplotlib import pyplot as plt
```

```python
DATA_DIR = Path('../data/data-for-processing')
CLEANED_DATA_LIST = list(DATA_DIR.glob('cleaned_*.feather'))
print(CLEANED_DATA_LIST)
```

```python
df = pd.read_feather(CLEANED_DATA_LIST[0])
df.columns
```

```python
for file_path in CLEANED_DATA_LIST:
    print('%s'%(file_path,))
    print('-'*30)
    df = pd.read_feather(file_path)
    print('-'*30)
    print(df['airport_jfk'].value_counts())
    print('-'*30)
    print(df['airport_lga'].value_counts())
    print('-'*30)
    print(df['airport_ewr'].value_counts())
    print('='*30)
```

```python
for file_path in CLEANED_DATA_LIST:
    print('%s'%(file_path,))
    print('-'*30)
    df = pd.read_feather(file_path)
    print(df['pickup_dropoff_distance'].value_counts())
    print('='*30)
```
